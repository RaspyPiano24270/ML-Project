# Model Training Notebook

This notebook runs the CLI training scripts directly from Python.
You can switch between:
- Baseline MLP model
- LSTM model

Make sure your CSV is inside the `data/` folder before running.

```python
import sys, subprocess

# Adjust these values to match your dataset
csv = 'data/rain.csv'        #  CHANGE IF NEEDED
target = 'rainfall_mm'       #  CHANGE IF NEEDED
seq_len = '24'
epochs = '5'

print('Settings Loaded:')
print('CSV     :', csv)
print('Target  :', target)
print('Seq_len :', seq_len)
print('Epochs  :', epochs)
```

##  Train Baseline MLP

```python
subprocess.run([
    sys.executable, 'models/baseline_model.py',
    '--csv', csv,
    '--target', target,
    '--seq_len', seq_len,
    '--epochs', epochs
], check=False)
print('Baseline training finished.')
```

## Train LSTM Model

```python
subprocess.run([
    sys.executable, 'models/lstm_model.py',
    '--csv', csv,
    '--target', target,
    '--seq_len', seq_len,
    '--epochs', epochs
], check=False)
print('LSTM training finished.')
```
