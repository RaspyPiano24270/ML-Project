```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model
from sklearn.preprocessing import MinMaxScaler
from data_loader import load_kaggle_csv

# Load Model
model = load_model('../models/lstm_regressor.h5')

# Load & Scale Data
df = load_kaggle_csv('../data/rain_prediction_dataset.csv')
values = df.values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(values)

# Grab a random 14-day window
import random
start_idx = random.randint(0, len(scaled_data) - 15)
sample_input = scaled_data[start_idx : start_idx+14] # Shape (14, 7)

# Reshape for LSTM (1, 14, 7)
sample_input = sample_input.reshape(1, 14, 7)

# Predict
prediction = model.predict(sample_input)
print(f"Model Prediction (Scaled): {prediction[0][0]:.4f}")
```
