# Exploratory Data Analysis (EDA)

Adjust the CSV path below depending on where your dataset is saved inside the `data/` folder.
This notebook allows you to quickly inspect column names, basic statistics, missing values, and distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = '../data/rain.csv'  # CHANGE THIS
df = pd.read_csv(csv_path)
df.head()
```

```python
# Shape (rows, columns)
df.shape
```

```python
# Column names
df.columns.tolist()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Plot distributions for all numeric features
df.hist(figsize=(12, 12))
plt.tight_layout()
plt.show()
```
